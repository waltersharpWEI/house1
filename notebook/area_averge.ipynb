```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
X = pd.read_csv('House_info_data.csv',encoding='gbk')
Y = pd.read_csv('House_price_data.csv',encoding='gbk')
```

```python
Y.head()
```

```python
X.loc[X.PPTY_MGMT_TYPE == u'公寓','OWNERSHIP_PERIOD'] = '40年'
X.loc[X.PPTY_MGMT_TYPE == u'普通住宅','OWNERSHIP_PERIOD'] = '70年'
X.loc[X.OWNERSHIP_PERIOD == u'40年','PPTY_MGMT_TYPE'] = '公寓'
X.loc[X.OWNERSHIP_PERIOD == u'70年','PPTY_MGMT_TYPE'] = '普通住宅'
data = X.loc[:,['HOUSE_SRN','CITY','AREA','OWNERSHIP_PERIOD','PPTY_MGMT_TYPE']]
data.dropna(how='any')
data.head()
```

```python
Y.HOUSE_SRN = Y.HOUSE_SRN.astype(str)
total = pd.merge(data,Y,on="HOUSE_SRN",how='left')
total['DATA_DATE'] = pd.to_datetime(total['DATA_DATE'])
total['DATA_DATE'] = (total.DATA_DATE - pd.to_datetime('2014/01/01')).astype('timedelta64[M]') + 1
total.head()
```

```python
total = total[total.AVG_PRICE.isnull() == False]
total = total[total['DATA_DATE'] > 0]
total = total[total['OWNERSHIP_PERIOD'] == u'70年']
X = total.loc[:,['CITY','AREA','OWNERSHIP_PERIOD','PPTY_MGMT_TYPE','DATA_DATE','AVG_PRICE']].dropna(how='any')
```

```python
X[X['AREA'] == u'静安']#[X['DATA_DATE'] ]
```

```python
samp = X[X['AREA'] == u'朝阳'].groupby(['DATA_DATE']).mean().reset_index()
length = samp.shape[0]
print length
a = samp.DATA_DATE.values.reshape(length, 1)
b = samp.AVG_PRICE.values.reshape(length, 1)
lr = LinearRegression()
lr.fit(a,b)

samp = X[X['AREA'] == u'静安'].groupby(['DATA_DATE']).mean().reset_index()
length = samp.shape[0]
print length
a1 = samp.DATA_DATE.values.reshape(length, 1)
b1 = samp.AVG_PRICE.values.reshape(length, 1)
lr1 = LinearRegression()
lr1.fit(a1,b1)


samp = X[X['AREA'] == u'福田'].groupby(['DATA_DATE']).mean().reset_index()
length = samp.shape[0]
print length
a2 = samp.DATA_DATE.values.reshape(length, 1)
b2 = samp.AVG_PRICE.values.reshape(length, 1)
lr2 = LinearRegression()
lr2.fit(a2,b2)

samp = X[X['AREA'] == u'天河区'].groupby(['DATA_DATE']).mean().reset_index()
length = samp.shape[0]
print length
a3 = samp.DATA_DATE.values.reshape(length, 1)
b3 = samp.AVG_PRICE.values.reshape(length, 1)
lr3 = LinearRegression()
lr3.fit(a3,b3)


plt.title(u'70 years OWNERSHIP_PERIOD')
plt.ylabel('average price')
plt.xlabel('months from 2014/01/01')
plt.scatter(a,b,color = 'black')
plt.plot(a, lr.predict(a),'k-',label='$peking chaoyang$')

plt.scatter(a1,b1,color='red')
plt.plot(a1, lr1.predict(a1),'r-',label='$shang hai jing an$')

plt.scatter(a2,b2,color='blue')
plt.plot(a2, lr2.predict(a2),'b-',label='$shen zhen fu tian$')

plt.scatter(a3,b3,color='yellow')
plt.plot(a3, lr3.predict(a3),'y-',label='$guang zhou tian he$')

plt.legend()  
```
